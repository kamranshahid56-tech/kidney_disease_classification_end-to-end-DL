```python
from ensure import ensure_annotations
```

```python
def product(x: int, y: int) -> int:
    return x * y    
```

```python
product(x = 2, y = "8")
```

```python
@ensure_annotations
def products(x: int, y: int) -> int:
    return x * y    
```

```python
products(x = 2, y = 8)
```
